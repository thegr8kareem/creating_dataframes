# Creating Dataframes


```python
import pandas as pd
import numpy as np

# df = dataframe

my_dict = {}
question1 = int(input('How many columns do you want to use: '))

df_values = []
extended_df_values = []


def list_gen():
    word = 'value'
    for j in range(question1):
        num = str(j+1)
        data = word + num
        # print(data)
        df_values.append(data)
        df_values[j] = []


list_gen()


j = 0
question3 = int(input('How many values are in each column: '))


for i in range(question1):
    question2 = input(f'Enter the name of column {i+1}: ')
    key = question2

    for i in range(question3):
        question4 = input(
            f'Enter the content(s) of the column {i + 1},one at a time: ')
        df_values[j].append(question4)

    my_dict.update({key: df_values[j]})
    j += 1


new_df = pd.DataFrame(my_dict)
new_df

question5 = input('Do you wish to add more columns to your data: ').capitalize()
question6 = int(input('How many do you want to add: '))

def list_gen2():
    word = 'value'
    for m in range(question6):
        num = str(m)
        data = word + num
        # print(data)
        extended_df_values.append(data)
        extended_df_values[m] = []

# This is for extension of the dataframe
def extended_df():
    m = 0
    for i in range(question6):
        question2 = input(f'Enter the name of column {i+1}: ')
        new_key = question2

        for j in range(question3):
            question4 = input(
                f'Enter the content(s) of the column {i + 1},one at a time: ')
            extended_df_values[m].append(question4)
        my_dict[new_key] = extended_df_values[m]
        m += 1



# Adding more columns to already created dataframe
if question5 == 'Yes':
    list_gen2()
    extended_df()
else:
    pass

new_df = pd.DataFrame(my_dict)
new_df

```
